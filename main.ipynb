```python
import matplotlib.pyplot as plt
from sympy import plot_implicit, symbols, Eq, And

from sympy import plot
from sympy.plotting.plot import Plot
from sympy import Symbol

from sympy import symbols, sin, cos
```

```python
x, y = symbols('x y')
p1 = plot_implicit(Eq(x**2 + y**2, 5), backend='matplotlib')
plt.show()
```

```python
x = symbols('x')
y1 = sin(x)
y2 = cos(x)

p = Plot((y1, (x, 0, 2)), (y2, (x, 0, 2)), title='Sine and cosine functions')

# Access the first series in the _series list and set its label to 'sin(x)'
p._series[0].label = 'sin(x)'

# Access the second series in the _series list and set its label to 'cos(x)'
p._series[1].label = 'cos(x)'

# Add a legend to the plot
p.legend = True

p.show()
```
